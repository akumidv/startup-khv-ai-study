```python
import numpy as np
from numpy.random import default_rng
import plotly.graph_objs as go
import pandas as pd

rng = default_rng()

step_n = 2000

price_close = []
price_open = []
price_high = []
price_low = []
prev_price = 5750
for idx in range(step_n):
    price_close.append(prev_price + prev_price * rng.choice([0.005, 0.01, 0.015, 0.02, 0.025, 0.03]) * rng.choice([-1, 1]) )
    prev_price = price_close[-1]
    price_open.append(prev_price + prev_price * rng.choice([0.005, 0.01, 0.015, 0.02, 0.025, 0.03]) * rng.choice([-1, 1]))
    prev_price = max(price_close[-1], price_open[-1])
    price_high.append(prev_price + prev_price * rng.choice([0.005, 0.01, 0.015, 0.02, 0.025, 0.03]))
    prev_price = min(price_close[-1], price_open[-1])
    price_low.append(prev_price - prev_price * rng.choice([0.005, 0.01, 0.015, 0.02, 0.025, 0.03]))
    prev_price = price_open[-1]

date = pd.date_range(start='1/1/2020', periods=step_n, freq='H')

fig = go.Figure()
fig.add_trace(go.Candlestick(x=date,
                open=price_open,
                high=price_high,
                low=price_low,
                close=price_close))
fig.show()
```
